```python
train.store_and_fwd_flag = pd.get_dummies(train.store_and_fwd_flag, drop_first=True)
train.vendor_id = pd.get_dummies(train.vendor_id, drop_first=True)
```

```python
train.pickup_datetime = pd.to_datetime(train.pickup_datetime)
train.dropoff_datetime = pd.to_datetime(train.dropoff_datetime)
```

```python
train['pickup_dow'] = train.pickup_datetime.dt.dayofweek
train['dropoff_dow'] = train.dropoff_datetime.dt.dayofweek
```
